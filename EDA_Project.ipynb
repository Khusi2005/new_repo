```python
df['Days']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df['Dep_Hours']=df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Minutes']=df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
del df['Dep_Time']
```

```python
df['Arrival_Time'].str.split(' ')
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Hours']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Minutes']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.head()
```

```python
del df['Arrival_Time']
```

```python
df.head()
```

```python
df['Duration'].unique()
```

```python
df['Duration']=df['Duration'].replace('5m','0h 5m')
```

```python
df['Duration'].unique()
```

```python
df['Duration'].value_counts()
```

```python
df.head()
```

```python
df['DH']=df['Duration'].str.split(' ').str[0].str.split('h').str[0].astype(int)*60
df['DM']=df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0).astype(int)
```

```python
df.head()
```

```python
df['Total_Duration']=df['DH']+df['DM']
```

```python
df.head()
```

```python
del df['Duration']
```

```python
del df['DH']
del df['DM']
```

```python
df.head()
```

```python
del df['Additional_Info']
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].value_counts()
```

```python
df['Total_Stops'] = df['Total_Stops'].str.replace('non-stop', '0')
df['Total_Stops'] = df['Total_Stops'].str.replace('1 stop', '1')
df['Total_Stops'] = df['Total_Stops'].str.replace('2 stops', '2')
df['Total_Stops'] = df['Total_Stops'].str.replace('3 stops', '3')
df['Total_Stops'] = df['Total_Stops'].str.replace('4 stops', '4')
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
df.head()
```

compare the mean price of every airline

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mean_price=df.groupby('Airline')['Price'].mean()
plt.figure(figsize=(15,5))
sns.barplot(x=mean_price.index,y=mean_price.values)
plt.title('Mean Price of Every Airline')
plt.xlabel('Airline')
plt.ylabel('Mean Price')
plt.show()
```

Average price vs no.of stops

```python
avg_price_stops=df.groupby('Total_Stops')['Price'].mean().reset_index()
sns.barplot(data=avg_price_stops,x='Total_Stops',y='Price')
plt.title('Average price vs no.of stops')
plt.xlabel('Total_Stops')
plt.ylabel('Price')
```

 Duration vs Price

```python
sns.scatterplot(data=df,x='Total_Duration',y='Price')
plt.show()
```

Distribution of Price

```python
plt.figure(figsize=(15,5))
sns.histplot(data=df,x='Price',bins=50)
plt.show()
```
